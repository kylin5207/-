2. 划分数据集

```python
X, y = df_wine.iloc[:,1:].values, df_wine.iloc[:,0].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

3. 标准化数据

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

4. 特征选择

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression(penalty='l1', C=0.1)
```

```python
lr.fit(X_train, y_train)
```

```python
lr.score(X_train, y_train)
```

```python
lr.score(X_test, y_test)
```

当逻辑回归应用于多分类任务上时，他使用OvR方法。其中，第一个截距项为类别1相对于2和3的结果，依次类推。

```python
lr.intercept_
```

```python
lr.coef_
```

```python
lr.coef_.shape
```

每一组包含13个权重值

```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure()
```

```python
colors=['blue', 'green', 'cyan', 'magenta', 'yellow', 'black', 'pink', 'lightgreen', 'lightblue', 'gray', 'indigo', 'orange']
```

```python
weights, params = [], []
```

```python
C = [0.001, 0.01, 0.1, 1, 10, 100]
```

```python
for c in C:
    lr = LogisticRegression(penalty='l1', C=c, random_state=0)
    lr.fit(X_train, y_train)
    weights.append(lr.coef_[1])
    params.append(c)
```

```python
weights = np.array(weights)
```

```python
weights
```

```python
params
```

```python
ax = plt.subplot(111)
for column, color in zip(range(weights.shape[1]), colors):
    plt.plot(params, weights[:, column], label=df_wine.columns[column+1], color=color)

plt.axhline(0, color='k', linestyle='--', linewidth=3)
plt.xlim([10**(-3), 10**2])
plt.ylabel("weight coefficient")
plt.xlabel("C")
plt.xscale('log')
plt.legend()
ax.legend(bbox_to_anchor(1.39, 1.03), ncol=1, fancybox=True)
plt.show()
```
